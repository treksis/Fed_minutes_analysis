```python
import requests
from bs4 import BeautifulSoup
import re

URL = "https://www.federalreserve.gov/monetarypolicy/fomcminutes20210616.htm"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
article = soup.find("div", {"id": "article"})
text = article.get_text()

#text = article.text

#letters_only = re.sub("[^a-zA-Z]",  # Search for all non-letters
#                          " ",          # Replace all non-letters with spaces
#                          str(text))
print(text)
```

```python
import nltk
from pprint import pprint

words = [w for w in text if w.isalpha()]
stopwords = nltk.corpus.stopwords.words("english")
words = [w for w in words if w.lower() not in stopwords]
pprint(nltk.word_tokenize(text), width=79, compact=True)
```

```python
words: list[str] = nltk.word_tokenize(text)
fd = nltk.FreqDist(words)

print(words)
```

```python
#fd.most_common(3)
fd.tabulate(3)
```

```python
lower_fd = nltk.FreqDist([w.lower() for w in fd])
```

```python
text = nltk.Text(words)
```

```python
text.concordance("participants", lines=10)
```

```python
concordance_list = text.concordance_list("participants", lines=2)
for entry in concordance_list:
    print(entry.line)
```

```python
finder = nltk.collocations.TrigramCollocationFinder.from_words(words)
```

```python
finder.ngram_fd.most_common(2)
```

```python
finder.ngram_fd.tabulate(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

from yellowbrick.text import FreqDistVisualizer
from yellowbrick.datasets import load_hobbies

# Load the text data
corpus = load_hobbies()

vectorizer = CountVectorizer()
docs       = vectorizer.fit_transform(corpus.data)
features   = vectorizer.get_feature_names()

visualizer = FreqDistVisualizer(features=features, orient='v')
visualizer.fit(docs)
visualizer.show()
```
