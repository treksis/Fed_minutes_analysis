```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
from collections import defaultdict

def extract_text(URL):
    
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article = soup.find("div", {"id": "article"}).findAll("p")
    
    data = defaultdict(list)
    current_chapter = ""

    for p in article:
        strong = p.find('strong')
        #print(strong)
        if strong is not None:
            current_chapter = strong.text
        else:
            data[current_chapter].append(p.text)
        
    return data

#for k,v in data.items():
#    print(f'{k} -> {v}')

extract_text("https://www.federalreserve.gov/monetarypolicy/fomcminutes20210728.htm").keys()
#extract_text("https://www.federalreserve.gov/monetarypolicy/fomcminutes20210616.htm")
```

```python
from bs4 import BeautifulSoup
  
# HTML Document
HTML_DOC = """
              <html>
               <head>
                   <title> Find Next Siblings </title>
               </head>
               <body>
                    <p class = "languages">1957: FORTRAN</p>
  
                    <p class = "languages">1972: C</p>
  
                    <p class = "languages">1983: C++</p>
  
                    <p class = "languages">1991: Python</p>
  
                    <p class = "languages">1993: Ruby</p>
  
                    <p class = "languages">1995: Java</p>
  
                    <p class = "languages">1995: PHP</p>
  
                    <p class = "languages">1995: JavaScript</p>
  
               </body>
             </html>
            """
  
# Function to find all the next siblings
def findNextSiblings(html):
  
    # parse html content
    soup = BeautifulSoup(html, "html.parser")
  
    element = soup.p
  
    # Extracting all the next siblings of an element
    nextSiblings = element.find_next_siblings("p")
  
    # Printing all the next siblings
    for nextSibling in nextSiblings:
        print(nextSibling)
  
  
# Function Call
findNextSiblings(HTML_DOC)
```

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.federalreserve.gov/monetarypolicy/fomcminutes20210728.htm'
soup = BeautifulSoup(requests.get(url).content, 'html.parser')

def get_paragraphs(first_paragraph):
    current_paragraph = first_paragraph
    while True:
        yield current_paragraph

        current_paragraph = current_paragraph.find_next('p')
        if not current_paragraph or (current_paragraph.strong and current_paragraph.strong.text.strip() != ''):
            break

for p in get_paragraphs(soup.select_one('p:contains("Discussion of Asset Purchases")')):
    print(p)
```

```python
def get_header(URL):
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article = soup.find("div", {"id": "article"}).findAll("strong")
    return article
```

```python
headers = get_header('https://www.federalreserve.gov/monetarypolicy/fomcminutes20210728.htm')

for header in headers:
    text=header.text
    print(text)
```
